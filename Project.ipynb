```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# Define models
logistic_regression = LogisticRegression(max_iter=1000)
random_forest = RandomForestClassifier()
svm_classifier = SVC()
gradient_boosting = GradientBoostingClassifier()
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import classification_report
# Train and evaluate models using cross-validation

# Perform cross-validation predictions
y_pred_log = cross_val_predict(logistic_regression, X_resampled, y_resampled, cv=5)
# Generate classification report
report = classification_report(y_resampled, y_pred_log)
print("Logistic Regression")
print(report)
```

```python
#random Forst
y_pred_random = cross_val_predict(random_forest, X_resampled, y_resampled, cv=5)
report = classification_report(y_resampled, y_pred_random)
print("Random Forest")
print(report)
```

```python
#Gradient Boosting
y_pred_gb = cross_val_predict(gradient_boosting, X_resampled, y_resampled, cv=5)
report = classification_report(y_resampled, y_pred_gb)
print("Gradient Boosting")
print(report)
```

```python
from sklearn.model_selection import GridSearchCV

# Hyperparameter grid for each model
param_grid_logistic = {'C': [0.1, 1, 10, 100]}
param_grid_rf = {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20]}
param_grid_gb = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}
```

```python
# Grid search for hyperparameter tuning
grid_search_logistic = GridSearchCV(LogisticRegression(max_iter=1000), param_grid_logistic, cv=5)
grid_search_rf = GridSearchCV(RandomForestClassifier(), param_grid_rf, cv=5)
#grid_search_svm = GridSearchCV(SVC(), param_grid_svm, cv=5)
grid_search_gb = GridSearchCV(gradient_boosting, param_grid_gb, cv=5)


# Fit grid search models
grid_search_logistic.fit(X_resampled, y_resampled)
grid_search_rf.fit(X_resampled, y_resampled)
#grid_search_svm.fit(X_resampled, y_resampled)
grid_search_gb.fit(X_resampled, y_resampled)

# Get best hyperparameters
best_params_logistic = grid_search_logistic.best_params_
best_params_rf = grid_search_rf.best_params_
#best_params_svm = grid_search_svm.best_params_
best_params_gb = grid_search_gb.best_params_

print("Best Hyperparameters for Logistic Regression:", best_params_logistic)
print("Best Hyperparameters for Random Forest:", best_params_rf)
#print("Best Hyperparameters for SVM:", best_params_svm)
print("Best hyperparameters for Gradient Boosting:", best_params_gb)
```

```python
#Implementing ADA BOOST
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
#Using hyper parameters
logistic_regression = LogisticRegression(C=100, max_iter=1000)
random_forest = RandomForestClassifier(max_depth = 5, n_estimators = 100)
#svm_classifier = SVC(C=100, gamma='auto')
gradient_boosting = GradientBoostingClassifier(learning_rate = 0.1, max_depth = 3, n_estimators = 50)

# Using Adaboost to adjust weights of the Logistic regression model
ada_clf_lr = AdaBoostClassifier(estimator=logistic_regression, n_estimators=50, learning_rate=1.0, random_state=42)
y_pred_ada_lr = cross_val_predict(ada_clf_lr, X_resampled, y_resampled, cv=5)
report = classification_report(y_resampled, y_pred_ada_lr)
print("Logistic Regression with Ada boost")
print(report)
print()
```

```python
#Random forest
ada_clf_rf = AdaBoostClassifier(estimator=random_forest, n_estimators=50, learning_rate=1.0, random_state=42)
y_pred_ada_rf = cross_val_predict(ada_clf_rf, X_resampled, y_resampled, cv=5)
report = classification_report(y_resampled, y_pred_ada_rf)
print("Random Forest with Ada boost")
print(report)
print()
```

```python
#Gradient Boosting
ada_clf_gb = AdaBoostClassifier(estimator=gradient_boosting, n_estimators=50, learning_rate=1.0, random_state=42)
y_pred_ada_gb = cross_val_predict(ada_clf_gb, X_resampled, y_resampled, cv=5)
report = classification_report(y_resampled, y_pred_ada_gb)
print("Gradient boosting with Ada boost")
print(report)
```

```python
# Implementing BAGGING
bagging_clf_lr = BaggingClassifier(estimator=ada_clf_lr, n_estimators=50, random_state=42)
y_pred_bagging_lr = cross_val_predict(bagging_clf_lr, X_resampled, y_resampled, cv=5)
report = classification_report(y_resampled, y_pred_bagging_lr)
print("Logistic Regression with bagging")
print(report)
```

```python
bagging_clf_rf = BaggingClassifier(estimator=ada_clf_rf, n_estimators=50, random_state=42)
y_pred_bagging_rf = cross_val_predict(bagging_clf_rf, X_resampled, y_resampled, cv=5)
report = classification_report(y_resampled, y_pred_bagging_rf)
print("Random Forest with bagging")
print(report)
```

```python
bagging_clf_gb = BaggingClassifier(estimator=ada_clf_gb, n_estimators=50, random_state=42)
y_pred_bagging_gb = cross_val_predict(bagging_clf_gb, X_resampled, y_resampled, cv=5)
report = classification_report(y_resampled, y_pred_bagging_gb)
print("Gradient boosting with bagging")
print(report)
```

```python
#Choosing the best classifier
#Random forest test
bagging_clf_rf.fit(X_resampled, y_resampled)
y_pred_random = bagging_clf_rf.predict(X_test)
print("Random Forest Test Set Performance:")
print(classification_report(y_test, y_pred_random))
```
